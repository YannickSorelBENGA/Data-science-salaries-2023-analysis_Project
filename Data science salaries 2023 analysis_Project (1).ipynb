### Information about Data_scientist salaries 2023 dataset

1. **Work_year:** This variable give information about the year of salary
2. **experience_level:** This variable likely categorizes employees into different levels of experience,  "Entry-level," "Experienced," "Mid-level," or "Senior-level." It provides information about an employee's career stage.

8. **remote_ratio:(Numeric Variable)** This variable represent the proportion  of remote work within an employee's job
9. **company_location:** This is likely the location or country where the employee's company is based.
10. **company_size:** This variable categorizes companies into different size categories, such as "Large," "Medium," or "Small."

* **Independance**
All the four group are independant

```python
entry_level = New_data[New_data['experience_level']=='Entry-level']['salary_in_usd']
Senior_level = New_data[New_data['experience_level']=='Senior-level']['salary_in_usd']
Experienced = New_data[New_data['experience_level']=='Experienced']['salary_in_usd']
Mid_level = New_data[New_data['experience_level']=='Mid-level']['salary_in_usd']
```

```python
stats.normaltest(Senior_level)
```

```python
stats.normaltest(Mid_level)
```

Since P value is less than 0.05, we reject H0, then dataset are not normal distributed

##### * We will use Krustal-Wallis Test to test if mediam growth is the same across the four groups

```python
stats.kruskal(entry_level,Senior_level,Experienced,Mid_level)
```

**In this case, the test statistic is 402.35 and the corresponding p-value is 6.85e-87. Since this p-value is less than 0.05, we can reject the null hypothesis, then There is  statistically association between experience level and salary.**

### Checking the Assumption of Regression

* **linearity** : The relationship between each predictor variable and the target variable is linear.<br>
* **Independence of Errors**: The errors (residuals) resulting from the model are independent of each other. In other words, the value of the error for one data point should not depend on the errors for other data points.<br>
* **Nomally Distributed**<br> 
* **Homoscedasticity** The variance of the errors is equal across all levels of the predictor variables.<br>


### a) Linearity

```python
# Residuals (errors) from the model
residuals = lr.resid

# Scatterplot of predicted values vs. residuals
plt.scatter(lr.fittedvalues, residuals)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Scatterplot of Predicted Values vs. Residuals')
plt.show()
```

### b) Independence of Errors

```python
# Fit a linear regression model
model = sm.OLS(y, X3).fit()

# Calculate the residuals (errors)
residuals = model.resid

# Perform the Durbin-Watson test
from statsmodels.stats.stattools import durbin_watson
durbin_watson_statistic = durbin_watson(residuals)

# Print the Durbin-Watson statistic
print(f'Durbin-Watson Statistic: {durbin_watson_statistic}')
```

Durbin-Watson statistic of approximately 1.8262 indicates that there is positive autocorrelation in the residuals of your linear regression model.
Positive autocorrelation means that the residuals are correlated in a positive direction, suggesting that the errors in our model are not completely independent.

### c) Normality

```python
fig = sm.qqplot(residuals, fit=True, line="45")
plt.show()
```

The data values in the plot fall along a roughly straight line at a 45-degree angle, Then data are normaly distributed

### d)Equal variances (homoscedasticity)

```python
# Fit a linear regression model
model = sm.OLS(y, X3).fit()

# Calculate the residuals (errors)
residuals = model.resid

# Get predicted values from the model
predicted_values = model.fittedvalues

# Create a scatterplot of residuals vs. predicted values
plt.scatter(predicted_values, residuals, c='blue', alpha=0.6)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='red', linestyle='--', label='Zero Residuals Line')
plt.title('Residuals vs. Predicted Values (Homoscedasticity Plot)')
plt.legend()
plt.grid(True)
plt.show()
```
